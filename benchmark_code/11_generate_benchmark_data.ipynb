```python
import numpy as np
from pathlib import Path


path_spec_data=Path.cwd().parent.parent/"spec_data"
```

```python
# Create insert file

import os
import msgpack
import pickle
from dynamic_entropy_search.convert_to_mgf import _write_spec
spec_num=[1000, 10000, 100000]
path_ms2=path_spec_data/"35_random_export_ms2"
path_output_spectrum=path_spec_data/"benchmark_spec"
path_output_spectrum.mkdir(parents=True, exist_ok=True)

for charge in [1, -1]:
    for num in spec_num:
        all_selected_spec=[]
        select_number=num//100
        
        for i in range(100):
            # Load the file
            with open(path_ms2/f"charge_{charge}/batch_{i}.bin", "rb") as f:
                all_spec=msgpack.load(f)
            
            all_spec=np.array(all_spec)
            # Select spectra randomly
            
            total_spectra_number=len(all_spec)
            
            selected=np.random.choice(total_spectra_number, select_number, replace=False)
            selected_spectra=all_spec[selected]
            all_selected_spec.append(selected_spectra)
        

        all_selected_spec=np.concatenate(all_selected_spec).tolist()

        # Write the spectra into the file
        with open(path_output_spectrum/f"insert_spectra-charge_{charge}-number_{num}.pkl", "wb") as fa,\
            open(path_output_spectrum/f"insert_spectra-charge_{charge}-number_{num}.mgf", "w") as fc:
            for spec in all_selected_spec:
                _write_spec(spec, spec["peaks"], fc)

            pickle.dump(all_selected_spec, fa)
        
        
```
